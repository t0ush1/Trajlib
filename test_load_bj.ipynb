```python
import pandas as pd
import numpy as np
import ast

path_data = pd.read_csv(
    "/home/wangzitong/Trajlib/resource/data/bj_roadmap_edge.geo",
    #names=["driver_id", "order_id", "timestamp", "lon", "lat"],
)

# with pd.option_context('display.max_colwidth', None):
#     print(path_data["coordinates"])

def coord_mean(coord_str):
    """
    coord_str 可能是字符串或者已经是列表。
    返回 Series: {'lon': 经度平均值, 'lat': 纬度平均值}
    """
    # 1. 解析为列表
    if isinstance(coord_str, str):
        coords = ast.literal_eval(coord_str)    # 例如 [[116.38..., 39.89...], [...]]
    else:
        coords = coord_str                      # 已经是列表就直接用
    
    # 2. 转成 numpy 数组并求均值
    arr = np.asarray(coords, dtype=float)       # 形状 (n, 2)
    lon_mean = arr[:, 0].mean()                 # 经度平均
    lat_mean = arr[:, 1].mean()                 # 纬度平均
    return pd.Series({'lon': lon_mean, 'lat': lat_mean})

# 新增两列 lon 和 lat
path_data[['lon', 'lat']] = path_data['coordinates'].apply(coord_mean)

print(path_data.head())
```

```python
import pandas as pd
import numpy as np
import ast

def norm_geo(x):
    """
    把各种 geo_id 统一成纯数字字符串：
      * 去空格、换行
      * 去小数点和尾随 '.0'
      * 去前导零（可选；如需保留前导零可删这行）
    """
    if pd.isna(x):
        return None
    s = str(x).strip()            # 去空白
    if s.endswith('.0'):          # 去尾随小数
        s = s[:-2]
    s = s.lstrip('0') or '0'      # 去前导零；若全是 0 保留一个 0
    return s


# ---------- 1⃣  处理 path_data ----------
path_data['geo_key'] = path_data['geo_id'].apply(norm_geo)

geo_lookup = (
    path_data
      .dropna(subset=['geo_key'])
      .set_index('geo_key')[['lon', 'lat']]
      .to_dict('index')                 # {'11467472': {'lon':..., 'lat':...}, ...}
)

# ---------- 2⃣  处理 raw_data ----------
def to_list(x):             # 把字符串形式的列表变成 list
    return ast.literal_eval(x) if isinstance(x, str) else x

for col in ['path', 'tlist', 'vflag']:
    raw_data[col] = raw_data[col].apply(to_list)

def align_vflag(row):       # 若 vflag 不是 list，复制成与 path 等长
    if not isinstance(row['vflag'], list):
        row['vflag'] = [row['vflag']] * len(row['path'])
    return row

raw_data = raw_data.apply(align_vflag, axis=1)

# ---------- 3⃣  explode 三列 ----------
trk_pts = (
    raw_data
      .assign(order_id=raw_data['id'], driver_id=raw_data['usr_id'])
      .explode(['path', 'tlist', 'vflag'], ignore_index=True)
      .rename(columns={'path': 'geo_id', 'tlist': 'timestamp'})
)

# ---------- 4⃣  标准化 geo_id → geo_key ----------
trk_pts['geo_key'] = trk_pts['geo_id'].apply(norm_geo)

# ---------- 5⃣  查哈希表添经纬度 ----------
trk_pts['lon'] = trk_pts['geo_key'].map(lambda k: geo_lookup.get(k, {}).get('lon', np.nan))
trk_pts['lat'] = trk_pts['geo_key'].map(lambda k: geo_lookup.get(k, {}).get('lat', np.nan))

# ---------- 6⃣  生成结果 ----------
result = trk_pts[['order_id', 'driver_id', 'geo_id', 'lon', 'lat', 'timestamp', 'vflag']]

print(result)
result.to_csv("/home/wangzitong/Trajlib/trajlib/data/traj_BJ_201511.csv", index=False, header=False, encoding="utf-8")

print("轨迹点数量:", result.shape[0])
print("轨迹数量：", result['order_id'].nunique())
print("用户数量：", result['driver_id'].nunique())
```
